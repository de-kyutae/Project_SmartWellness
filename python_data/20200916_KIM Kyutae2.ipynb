## 영업중인 음식점만 필터링

```python
df_open = df[df['상세영업상태명'] == "영업"]
df_open
```

```python
df_open.isnull().sum()
```

## 비어 있는 주소란에 'NA'를 채워넣는다.

```python

df_filled = df_open.fillna('NA')
df_filled
```

```python
df_open_adress = df_filled[df_filled['소재지전체주소'] != 'NA']
df_open_adress
```

```python
df_open_adress.isnull().sum()
```

## 통닭집 찾기

```python
df_chicken = df_open_adress[df_open_adress.업태구분명 == ("호프/통닭" or "통닭(치킨)")]
df_chicken
```

## 주소형식 변경

```python
df_chicken['지역구'] = df_chicken.소재지전체주소.str.split(' ').str[1]
df_chicken['지역구']
```

```python
df_chicken['세부주소'] = df_chicken.소재지전체주소.str.split(' ').str[2]
df_chicken['세부주소']
```

```python
df_chicken_trans = df_chicken.T

df_chicken_trans
```

```python
df_chicken_trans.index
```

## 지역구별로 구분

```python
location_chicken = df_chicken.groupby(['지역구']).count()
location_chicken['번호']
```
