```python
###concatenate()
# - 지정한 방향으로 배열 결합
# - axis = 0 : 열방향
# - axis = 1 : 행방향
```

```python
import numpy as np
A = np.array([[1,2],[3,4]])
B = np.array([[1,0],[0,1]])

C=np.concatenate((A,B), axis=0) #열방향
print(C)
print()
D=np.concatenate((A,B), axis=1) #행방향
print(D)
```
