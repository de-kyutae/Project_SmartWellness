## 계층적 색인 지정

```python
# 새로운 데이터
f = open("csv_mindex.csv", 'w')
data = """
key1,key2,value1,value2
one,a,1,2
one,b,3,4
one,c,5,6
one,d,7,8
two,a,9,10
two,b,11,12
two,c,13,14
two,d,15,16
"""
f.write(data)
f.close()
```

```python
#계층적 색인 지정
parsed = pd.read_csv("csv_mindex.csv", index_col=['key1','key2'])
parsed
```

```python
f = open("ex3.txt", 'w')
data = """
            A         B         C
aaa -0.264438 -1.026059 -0.619500
bbb  0.927272  0.302904 -0.032399
ccc -0.264273 -0.386314 -0.217601
ddd -0.871859 -0.348382  1.100491
"""
f.write(data)
f.close()
```

```python
result = pd.read_table('ex3.txt', sep='\s+') #열 라인에 맞도록 구분하기 위한 옵션
result
```

```python
f = open("ex4.csv", 'w')
data = """#Hey!
a,b,c,d,message
#Hello puthon
#CSV file
1,2,3,4,hello
5,6,7,8,world
9,10,11,12,foo
"""

f.write(data)
f.close()

#제대로 파일을 읽어 들일수가 없다.
```

```python
f = open("ex5.csv", 'w')
data = """something,a,b,c,d,message
one,1,2,3,4,NA
two,5,6,,8,world
three,9,10,11,12,foo
"""
f.write(data)
f.close()
```

```python
result = pd.read_csv('ex5.csv')
result
```
