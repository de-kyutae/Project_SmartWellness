```python
import pandas as pd
import numpy as np
```

```python
df_hospital = pd.read_excel("data/6270000/6270000_대구광역시_01_01_01_P_병원.xlsx")
df_general = pd.read_excel("data/6270000/6270000_대구광역시_01_01_02_P_의원.xlsx")
```

```python
df_hospital = df_hospital.copy()
df_general = df_general.copy()
```

```python
titles_hospital = df_hospital.columns
titles_hospital
```

```python
title_general = df_general.columns
title_general
```

## 열 필터링

```python
df_hospital_list = df_hospital[['상세영업상태명','소재지전체주소','도로명전체주소','사업장명','업태구분명']]
df_hospital_list
```

```python
df_hospital_list.info()
```

```python
df_general_list = df_general[['상세영업상태명','소재지전체주소','도로명전체주소','사업장명','업태구분명']]
df_general_list
```

## 영업중만 남김

```python
df_hospital = df_hospital_list[df_hospital_list['상세영업상태명']=='영업중']
df_hospital
```

```python
df_general = df_general_list[df_general_list['상세영업상태명']=='영업중']
df_general
```

## 주소지 공백 삭제

```python
df_hospital['도로명전체주소'] = df_hospital['도로명전체주소'].str.lstrip()
df_hospital
```

```python
df_hospital['소재지전체주소'] = df_hospital['소재지전체주소'].str.lstrip()
df_hospital
```

```python
df_general['도로명전체주소'] = df_general['도로명전체주소'].str.lstrip()
df_general
```

```python
df_general['소재지전체주소'] = df_general['소재지전체주소'].str.lstrip()
df_general
```

## 자료 합치기

```python
df = pd.concat([df_hospital,df_general], ignore_index=True)
df
```

## 병원, 의원 데이터

```python
df_h = df[df['업태구분명'].isin(['병원','종합병원','의원'])]
df_h
```

## 치과 데이터

```python
df_t = df[df['업태구분명'].isin(['치과의원','치과병원'])]
df_t
```

## 결측치 삭제

```python
# 병원 데이터
df_h = df_h.dropna(axis=0)
df_h.info()
```

```python
# 치과 데이터
df_t = df_t.dropna(axis=0)
df_t.info()
```

## 주소 분류

```python
df_h[['시도','구','동','나머지']] = df_h['도로명전체주소'].str.split(' ', n=3, expand=True)
df_h.head()
```

```python
df_t[['시도','구','동','나머지']] = df_t['도로명전체주소'].str.split(' ', n=3, expand=True)
df_t.head()
```

## count 열 만들기

```python
#병원 데이터
df_h['count'] = 1
df_h.head()
```

```python
#치과 데이터
df_t['count'] = 1
df_t.head()
```

## 구별 병원 수

```python
# 병원
hospital = df_h.groupby('구').sum()
hospital = hospital.sort_values(by='count', ascending=False)
hospital
```

```python
# 치과
dental = df_t.groupby('구').sum()
dental = dental.sort_values(by='count', ascending=False)
dental
```
